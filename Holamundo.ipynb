```python
# importarmos las librerias: numpy y matplotlib
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Definimos 2 arreglos de dimensiones 100
ys = 200 + np.random.randn(100)
x = np.arange (0, 100,1)
```

```python
#Grsficamos los arreglos en un plano 2D
plt.plot(x,ys,'*')
plt.fill_between(x, ys, 195,facecolor='r', alpha=0.6)

#agregamos titulo a la grafica
plt.title("Sample Visualization")
plt.xlabel("eje de las x", size =16)
plt.ylabel("eje de las y", size =16)
plt.show()
```
